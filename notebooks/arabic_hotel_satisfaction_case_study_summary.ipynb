# Arabic Hotel Reviews Satisfaction Analysis with Machine Learning

This notebook summarizes a portfolio case study for analyzing Arabic hotel customer reviews and predicting customer satisfaction from review text.

## Objective

The goal is to show how Arabic customer feedback can be converted into business insights: satisfaction distribution, hotel-level comparisons, stay-length analysis, and machine learning classification.

## Dataset

The analysis uses the public HARD Arabic Hotel Reviews Dataset. The dataset includes Arabic review text, ratings, sentiment labels, hotel identifiers, user identifiers, and stay length.

The public dataset does not include a review date column, so time-trend claims are intentionally excluded.

```python
import pandas as pd

metrics = pd.DataFrame({
    'metric': ['accuracy', 'precision', 'recall', 'f1', 'roc_auc'],
    'value': [0.8899, 0.8768, 0.9073, 0.8918, 0.9525]
})
metrics
```

## Executive Results

- Reviews after cleaning: 93,631
- Hotels covered: 976
- Average rating: 3.12 out of 5
- High satisfaction share: 50.01%
- Model accuracy: 88.99%
- ROC AUC: 95.25%
